```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#---Obtención de los datos---#
#---Obtaining data---#
Datos = pd.read_csv('captura pesquera nacional.csv')

#---Separar por litoral del Pacífico y litoral del Golfo---#
#---Separate by Pacific and Gulf coasts---#
Litoral_del_Pacifico = Datos[Datos['LITORAL'] == 'Litoral del Pacifico' ]
Litoral_del_Golfo_y_Mar_Caribe = Datos[Datos['LITORAL'] == 'Litoral del Golfo y Mar Caribe' ]

#---Graficar año vs Toneladas---#
#---Graph year vs. tons---#
ax = plt.gca()

plt.title('Captura Pesquera Nacional\n')
plt.ylabel('Toneladas')
plt.xlabel('AÑO')

Litoral_del_Golfo_y_Mar_Caribe.plot(kind='line',x='AÑO',y='CAPTURA_(TONELADAS)',ax=ax,label='Litoral del Golfo y Mar Caribe')
Litoral_del_Pacifico.plot(kind='line',x='AÑO',y='CAPTURA_(TONELADAS)', color='red', ax=ax, label='Litoral del Pacífico')
plt.grid()
plt.show()
```

```python
#---Obtener datos x,y de las costas---#
#---Obtain x,y data from the coasts---#
x_Pacifico = Litoral_del_Pacifico['AÑO']
y_Pacifico = Litoral_del_Pacifico['CAPTURA_(TONELADAS)']

x_Golfo = Litoral_del_Golfo_y_Mar_Caribe['AÑO']
y_Golfo = Litoral_del_Golfo_y_Mar_Caribe['CAPTURA_(TONELADAS)']
```

```python
#---Ajustar una recta y una curva de grado 2 a los datos---#
#---Fit a line and a grade 2 curve to the data---#
z = np.polyfit(x_Pacifico, y_Pacifico, 1)
p = np.poly1d(z)

p2 = np.poly1d(np.polyfit(x_Pacifico, y_Pacifico,2))

plt.title('Captura Pesquera Nacional\nLitoral del Pacífico')
plt.ylabel('Toneladas')
plt.xlabel('AÑO')

xp = np.linspace(1990, 2018, 100)
#_ = plt.plot(x_Pacifico, y_Pacifico, '.', xp, p(xp), '-', xp, p2(xp), '--')
plt.plot(x_Pacifico, y_Pacifico, '.',label='Producción por año')
plt.plot(xp, p(xp), '-',label='Modelo Lineal')
plt.plot(xp, p2(xp), '--',label='Modelo Cuadrático')

plt.legend()
plt.grid()
plt.show()
#---Guardar imagen---#
#---Save image---#
```

### Se observa que la cantidad de toneladas de peces que se capturan en el litoral del Pacífico tiende a aumentar al paso de los años, aunque en los más recientes años ha disminuido.
### It is observed that the amount of tons of fish caught on the Pacific coast tends to increase over the years, although in more recent years it has decreased.

```python
#---Ajustar una recta y una curva de grado 2 a los datos---#
#---Fit a line and a grade 2 curve to the data---#
z = np.polyfit(x_Golfo, y_Golfo, 1)
p = np.poly1d(z)

p2 = np.poly1d(np.polyfit(x_Golfo, y_Golfo,2))


plt.title('Captura Pesquera Nacional\nLitoral del Golfo y Mar Caribe')
plt.ylabel('Toneladas')
plt.xlabel('AÑO')

xp = np.linspace(1990, 2018, 100)
#_ = plt.plot(x_Golfo, y_Golfo, '.', xp, p(xp), '-', xp, p2(xp), '--')
plt.plot(x_Golfo, y_Golfo, '.',label='Producción por año')
plt.plot(xp, p(xp), '-',label='Modelo Lineal')
plt.plot(xp, p2(xp), '--',label='Modelo Cuadrático')

plt.legend()
plt.grid()
plt.show()
#---Guardar imagen---#
#---Save image---#
```

### Se observa que la cantidad de toneladas de peces que se capturan en el litoral del Pacífico tuvo una tendencia a disminuir desde el 95, alcanzando un mínimo entre los años 2007 y 2012, pero ha ido en aumento en los últimos años.
### It can be seen that the amount of tons of fish caught on the Pacific coast had a tendency to decrease since 1995, reaching a minimum between 2007 and 2012, but has been increasing in recent years.
